```julia
using Gadfly
using Interact
```

```julia
set_default_plot_size(25cm, 15cm)
```

$$ U_A = x_1^A x_2^A \quad \quad U_B = x_1^B x_2^B$$

```julia
U_A(x1, x2) = x1*x2
U_B(x1, x2) = (ω1-x1)*(ω2-x2);
```

```julia
plot_U_A(x,x1,x2) = U_A(x1, x2)/x
plot_U_B(x,x1,x2) = U_B(x1, x2)/(x-ω1)+ω2;
```

```julia
ω1 = 10
ω2 = 20

@manipulate for x1 in 1:ω1-1, x2 in 1:ω2-1
    domain = linspace(0, ω1, 1000)
    lower_intercept = (U_A(x1, x2)/2)^(1/2)
    upper_intercept = (2ω1+ω2-((2ω1+ω2)^2-8*(ω1*ω2-U_B(x1,x2)))^(1/2))/4
    plot(
        # Initial State
        layer(x=[x1], y=[x2], Geom.point, Theme(default_color=colorant"black")),
        layer(x=domain, y=plot_U_A(domain, x1, x2), Geom.line, Theme(default_color=colorant"black")),
        layer(x=domain, y=plot_U_B(domain, x1, x2), Geom.line, Theme(default_color=colorant"black")),
        layer(x=domain, y=2*domain, Geom.line, Theme(default_color=colorant"black")),
        # Core
        layer(x=linspace(lower_intercept, upper_intercept, 1000), y=2*linspace(lower_intercept, upper_intercept, 1000), Geom.line, Theme(default_color=colorant"red", line_width=4pt)),
        # Setup
        Coord.Cartesian(xmin=0,xmax=ω1,ymin=0,ymax=ω2,fixed=true),
        Guide.xlabel("Good 1"),
        Guide.ylabel("Good 2")
        )
end
```

# Add Prices

```julia
B_x1(ω1_A, ω2_A, P1, P2) = (P1*ω1_A+P2*ω2_A)/(2P1)
B_x2(ω1_A, ω2_A, P1, P2) = (P1*ω1_A+P2*ω2_A)/(2P2)
B_line(x, ω1_A, ω2_A, P1, P2) = -(P1/P2)*x+(P1*ω1_A+P2*ω2_A)/P2;
```

```julia
ω1 = 210
ω2 = 210

@manipulate for ω1_A in 0:10:ω1-1, ω2_A in 0:10:ω2-1, P1 in 1:10, P2 in 1:10
    domain = linspace(0, ω1, 1000)
    lower_intercept = (U_A(ω1_A, ω2_A))^(1/2)
    upper_intercept = (ω1+ω2-((ω1+ω2)^2-4*(ω1*ω2-U_B(ω1_A, ω2_A)))^(1/2))/2
    
    p1 = plot(
        # Initial State
        layer(x=[ω1_A], y=[ω2_A], Geom.point),
        layer(x=domain, y=plot_U_A(domain, ω1_A, ω2_A), Geom.line),
        layer(x=domain, y=plot_U_B(domain, ω1_A, ω2_A), Geom.line),
        layer(x=domain, y=domain, Geom.line, Theme(default_color=colorant"black")),        
        # Core
        layer(x=linspace(lower_intercept, upper_intercept, 1000), y=linspace(lower_intercept, upper_intercept, 1000), Geom.line, Theme(default_color=colorant"red", line_width=4pt)),        
        # Exchange
        layer(x=[B_x1(ω1_A, ω2_A, P1, P2)], y=[B_x2(ω1_A, ω2_A, P1, P2)], Geom.point, Theme(default_color=colorant"black")),
        layer(x=[ω1-B_x1(ω1-ω1_A, ω2-ω2_A, P1, P2)], y=ω2-[B_x2(ω1-ω1_A, ω2-ω2_A, P1, P2)], Geom.point, Theme(default_color=colorant"black")),
        layer(x=domain, y=B_line(domain, ω1_A, ω2_A, P1, P2), Geom.line, Theme(default_color=colorant"black")),        
        layer(x=domain, y=plot_U_A(domain, B_x1(ω1_A, ω2_A, P1, P2), B_x2(ω1_A, ω2_A, P1, P2)), Geom.line, Theme(default_color=colorant"black")),
        layer(x=domain, y=plot_U_B(domain, ω1-B_x1(ω1-ω1_A, ω2-ω2_A, P1, P2), ω2-B_x2(ω1-ω1_A, ω2-ω2_A, P1, P2)), Geom.line, Theme(default_color=colorant"black")),
        # Setup
        Coord.Cartesian(xmin=0,xmax=ω1,ymin=0,ymax=ω2,fixed=true),
        Guide.xlabel("Good 1"),
        Guide.ylabel("Good 2"),
        Guide.Title("Edgeworth Box and Contract Curve")
        )
    
    p2 = plot(
        # UPF
        layer(x=[0 U_A(ω1,ω2)], y=[U_B(0,0) 0], Geom.line,  Theme(default_color=colorant"black")),
        # Rawls
        layer(x=[U_A(ω1,ω2)/2], y=[U_A(ω1,ω2)/2], Geom.point,  Theme(default_color=colorant"blue")),
        layer(x=[U_A(ω1,ω2)/2 50000], y=[U_A(ω1,ω2)/2 U_A(ω1,ω2)/2], Geom.line,  Theme(default_color=colorant"blue")),
        layer(x=[U_A(ω1,ω2)/2 U_A(ω1,ω2)/2], y=[U_A(ω1,ω2)/2 50000], Geom.line,  Theme(default_color=colorant"blue")),
        # CD
        layer(x=[U_A(ω1,ω2)/2], y=[U_A(ω1,ω2)/2], Geom.point,  Theme(default_color=colorant"red")),
        layer(x=1:50000, y=((U_A(ω1,ω2)/2)*(U_B(0,0)/2))./(1:50000), Geom.line,  Theme(default_color=colorant"red")),
        # Setup
        Coord.Cartesian(xmin=0,xmax=50000,ymin=0,ymax=50000, fixed=true),
        Guide.xlabel("Utility A"),
        Guide.ylabel("Utility B"),
        Guide.Title("Welfare and UPF")
        )
    
    
    hstack(p1,p2)
    
end
```
