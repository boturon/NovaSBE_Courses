# Data for each year

```python
dict_col_count = {}
for i in range(36):
    for col in set(pd.read_csv("/home/boturon/proj/git/incendios/data/incendios"+str(i+1980)+".csv", low_memory=False).columns):
        if col in dict_col_count:
            dict_col_count[col] +=1
        else:
            dict_col_count[col] = 1
```

```python
dict_col_count
```

```python
dict_col_years = {}
for i in range(36):
    for col in set(pd.read_csv("/home/boturon/proj/git/incendios/data/incendios"+str(i+1980)+".csv", low_memory=False).columns):
        if col in dict_col_years:
            dict_col_years[col].append(i+1980)
        else:
            dict_col_years[col] = [i+1980]
```
