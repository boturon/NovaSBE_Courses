# PPF Specific Factors Model

```python
def g1(input_M,input_L):
    return(np.sqrt(np.multiply(input_M,input_L)))
def f1(input_P,input_L):
    return(np.multiply(input_P,input_L))
```

## 1.b)

```python
M = 1; P = 1; L = np.linspace(0,16,100)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

# Before
## Economy's allocation of labor (AA)
ax.plot(-L,-L[-1]+L, label="Allocation of $L$", color="r")
## Production function of x
ax.plot(g1(M,L),-L, label=r"$g(M,L_x)$", color="b")
## Production function of y
ax.plot(-L,f1(P,L), label=r"$f(P,L_y)$", color="g")
## Production possibility frontier
ax.plot(g1(M,L),f1(P,L[-1]-L), label="PPF", color="grey")

ax.legend()
ax.axhline(0, color='k')
ax.axvline(0, color='k')
ax.set_xlim(-L[-1], L[-1])
ax.set_ylim(-L[-1], L[-1])
ax.annotate(r"$x$", (L[-1]-1,0-1))
ax.annotate(r"$y$", (0-1,L[-1]-1))
ax.annotate(r"$L_y$", (-(L[-1]-1),-(0-1)))
ax.annotate(r"$L_x$", (-(0-1),-(L[-1]-1)))
ax.grid()
```

# 3

```python
def X(T,L):
    return(np.sqrt(np.multiply(T,L)))
def Y(K,L):
    return(np.multiply(K,L))
def U(x,y):
    return(np.multiply(x,y))
```

```python
K = 4; T = 4; L = np.linspace(0,48,10000)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

# Before
## Economy's allocation of labor (AA)
ax.plot(-L,-L[-1]+L, label="Allocation of $L$", color="r")
## Production function of x
ax.plot(X(T,L),-L, label=r"$X(T,L_x)$", color="b")
## Production function of y
ax.plot(-L,Y(K,L), label=r"$Y(K,L_y)$", color="g")
## Production possibility frontier
ax.plot(X(T,L),Y(K,L[-1]-L), label="PPF", color="grey")
## Utility
ax.plot(X(T,np.linspace(0,100000,1000000)[1:]),1024/X(T,np.linspace(0,100000,1000000)[1:]), label="$U(x,y)$", color="pink")
## Equilibrium
ax.scatter(8,128, color="k")
ax.annotate(r"$(8,128)$", (18-8,128-8))

ax.legend()
ax.axhline(0, color='k')
ax.axvline(0, color='k')
ax.set_xlim(-L[-1], 200)
ax.set_ylim(-L[-1], 200)
ax.annotate(r"$x$", (200-8,0-8))
ax.annotate(r"$y$", (0-8,200-8))
ax.annotate(r"$L_y$", (-(L[-1]-1),-(0-2)))
ax.annotate(r"$L_x$", (-(0-1),-(L[-1]-1)))
ax.grid()
```
