Key equations:
$$ Y = A \cdot F(K,L) $$
$$ \Delta k/k = s \cdot (y/k) - s \delta -n $$

```julia
#function solow_swan()
    
    timer  = Observable(0.0)
    @async while true
        sleep(.5)
        timer[] = timer[]+0.1
    end
    
    K = 100
    
    @manipulate for A = 1:.5:5, s = 0:.1:1, n = 0:.1:1, δ = 0:.1:1, L = 50:150, t = timer
        global K
        Y = A*(K^.6)*(L^.4)
        Δk_k = s*(Y/K)-s*δ-n
        K = K*(1+Δk_k)
        k_axis = .001:1:100
        plot(x = k_axis, [(s*(A*k_axis.^.6)./k_axis) ones(100)*(s*δ+n)])
        vline!([K/L])
        ylims!((0,25))
        xlims!((0,25))

    end
#end
```

```julia
#function solow_swan()
    
    timer  = Observable(0.0)
    @async while true
        sleep(.5)
        timer[] = timer[]+0.1
    end
    
    K = 600

    @manipulate for A = 1:.5:5, s = 0:0.05:1, n = 0:0.01:0.1, δ = 0:0.05:1, L = 1:300, t = timer
        global K
        Y = A*(K^.6)*(L^.4)
        Δk_k = s*(Y/K)-s*δ-n
        K = K*(1+Δk_k)
    
        k_axis = 1:1:100
        
        plot(
            layer(x = k_axis, y = s*(A*k_axis.^.6)./k_axis, Geom.line),
            layer(x = k_axis, y = s*δ+n, Geom.line),
            layer(x = [K/L K/L], y = [(s*(A*(K/L)^.6)/(K/L)) s*δ+n]),
            Coord.Cartesian(xmin=0,xmax=10,ymin=0,ymax=1),
            Guide.xlabel("k = K/L (capital per worker)"),
            Guide.ylabel("Determinants of Δk/k")
        )
    end
#end
```
