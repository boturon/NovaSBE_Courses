```julia
using Gadfly, Interact
```

```julia
K1 = 1000; xs = 0:200; r0 = 0.5; ρ0 = 0; z2 = 2; Ω1 = 150; α = 0.7
```

```julia
plot(
    layer(x = xs, y = ((1+r0)/(1+ρ0)).*xs, Geom.line, Theme(default_color=colorant"black")),
    layer(x = xs, y = -(1+r0).*xs.+(1+r0)*Ω1, Geom.line, Theme(default_color=colorant"black")),
    layer(x = xs, y = z2.*max.(K1^α.+-xs,0).^α, Geom.line, Theme(default_color=colorant"black")),
    layer(x = xs, y = ℯ^(), Geom.line, Theme(default_color=colorant"black")),
    Coord.Cartesian(xmin=0,xmax=xs[end],ymin=0,ymax=xs[end],fixed=true),
    Guide.xlabel("C₁"), Guide.ylabel("C₂")
)
```
