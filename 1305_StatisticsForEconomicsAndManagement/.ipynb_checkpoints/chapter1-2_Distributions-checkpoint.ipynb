```julia
using Distributions
using DataFrames
using Interact
using Gadfly
```

```julia
set_default_plot_size(15cm, 15cm)
```

```julia
function discrete_sample_pdf(array_of_values)
    pdf_df = DataFrame(x_values = Int64[], y_values = Int64[])
    for x_value in sort(unique(array_of_values))
        counter = 0
        for value in array_of_values
            if value == x_value
                counter += 1
            end
        end
        push!(pdf_df, [x_value, counter])
    end
    return pdf_df
end
```

```julia
function discrete_sample_cdf(array_of_values)
    cdf_df = DataFrame(x_values = Int64[], y_values = Int64[])
    sum = 0
    for x_value in sort(unique(array_of_values))
        for value in array_of_values
            if value == x_value
                sum += 1
            end
        end
        push!(cdf_df, [x_value, sum])
    end
    return cdf_df
end
```

# Distributions

## Discrete

### with finite support

#### Bernoulli distribution

The Bernoulli distribution, which takes value 1 with probability p and value 0 with probability q = 1 − p.

$$ P(X = k) = \begin{cases}
1 - p & \quad \text{for } k = 0, \\
p & \quad \text{for } k = 1.
\end{cases} $$

```julia
bernoulli_values = rand(Bernoulli(.2),5)
```

```julia
@manipulate for trials in 0:2:1000, p in 0:.1:1
    xy_values_pdf = discrete_sample_pdf(rand(Bernoulli(p),trials))
    pl_pdf = plot(x=xy_values_pdf[:x_values], y=xy_values_pdf[:y_values]/trials, Geom.bar,Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))

    xy_values_cdf = discrete_sample_cdf(rand(Bernoulli(p),trials))
    pl_cdf = plot(x=xy_values_cdf[:x_values], y=xy_values_cdf[:y_values]/trials, Geom.bar,Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))
    
    vstack(pl_pdf, pl_cdf)
end
```

#### Binomial distribution

The binomial distribution, which describes the number of successes in a series of independent Yes/No experiments all with the same probability of success.

$$P(X = k) = {n \choose k}p^k(1-p)^{n-k},  \quad \text{ for } k = 0,1,2, \ldots, n.$$

```julia
binomial_values = rand(Binomial(10,.2),5)
```

```julia
@manipulate for trials in 0:10:10000, p in 0:.01:1, n in 0:10:1000
    xy_values_pdf = discrete_sample_pdf(rand(Binomial(n,p),trials))
    pl_pdf = plot(x=xy_values_pdf[:x_values], y=xy_values_pdf[:y_values]/trials, Geom.bar, Theme(bar_spacing=1mm))

    xy_values_cdf = discrete_sample_cdf(rand(Binomial(n,p),trials))
    pl_cdf = plot(x=xy_values_cdf[:x_values], y=xy_values_cdf[:y_values]/trials, Geom.bar, Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))
    
    vstack(pl_pdf, pl_cdf)
end

```

#### Discrete uniform distribution

The discrete uniform distribution, where all elements of a finite set are equally likely. This is the theoretical distribution model for a balanced coin, an unbiased die, a casino roulette, or the first card of a well-shuffled deck.

$$P(X = k) = 1 / (b - a + 1) \quad \text{for } k = a, a+1, \ldots, b.$$

```julia
discrete_uniform_values = rand(DiscreteUniform(0,10),5)
```

```julia
@manipulate for trials in 0:10:1000, a in 0:50, b in 51:100
    xy_values_pdf = discrete_sample_pdf(rand(DiscreteUniform(a,b),trials))
    pl_pdf = plot(x=xy_values_pdf[:x_values], y=xy_values_pdf[:y_values]/trials, Geom.bar, Theme(bar_spacing=1mm))

    xy_values_cdf = discrete_sample_cdf(rand(DiscreteUniform(a,b),trials))
    pl_cdf = plot(x=xy_values_cdf[:x_values], y=xy_values_cdf[:y_values]/trials, Geom.bar, Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))
    
    vstack(pl_pdf, pl_cdf)
end
```

### with infinite support

#### Geometric distribution

The geometric distribution, a discrete distribution which describes the number of attempts needed to get the first success in a series of independent Bernoulli trials, or alternatively only the number of losses before the first success (i.e. one less).

$$ P(X = k) = p (1 - p)^k, \quad \text{for } k = 0, 1, 2, \ldots.$$

```julia
geometric_values = rand(Geometric(.2),5)
```

```julia
@manipulate for trials in 0:10:1000, p in 0:.1:1
    xy_values_pdf = discrete_sample_pdf(rand(Geometric(p),trials))
    pl_pdf = plot(x=xy_values_pdf[:x_values], y=xy_values_pdf[:y_values]/trials, Geom.bar, Theme(bar_spacing=1mm))

    xy_values_cdf = discrete_sample_cdf(rand(Geometric(p),trials))
    pl_cdf = plot(x=xy_values_cdf[:x_values], y=xy_values_cdf[:y_values]/trials, Geom.bar, Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))
    
    vstack(pl_pdf, pl_cdf)
end
```

#### Poisson distribution

The Poisson distribution, which describes a very large number of individually unlikely events that happen in a certain time interval. 

$$P(X = k) = \frac{\lambda^k}{k!} e^{-\lambda}, \quad \text{ for } k = 0,1,2,\ldots.$$

```julia
poisson_values = rand(Poisson(2),5)
```

```julia
@manipulate for trials in 0:10:1000, λ in 0:.1:10
    xy_values_pdf = discrete_sample_pdf(rand(Poisson(λ),trials))
    pl_pdf = plot(x=xy_values_pdf[:x_values], y=xy_values_pdf[:y_values]/trials, Geom.bar, Theme(bar_spacing=1mm))

    xy_values_cdf = discrete_sample_cdf(rand(Poisson(λ),trials))
    pl_cdf = plot(x=xy_values_cdf[:x_values], y=xy_values_cdf[:y_values]/trials, Geom.bar, Coord.Cartesian(ymin=0,ymax=1), Theme(bar_spacing=1mm))
    
    vstack(pl_pdf, pl_cdf)
end
```

## Continuous

### supported on a bounded interval

#### Continuous uniform distribution

The continuous uniform distribution or rectangular distribution on [a,b], where all points in a finite interval are equally likely

$$ f(x; a, b) = \frac{1}{b - a}, \quad a \le x \le b$$

```julia
continuous_uniform_values = rand(Uniform(0,10),5)
```

```julia
@manipulate for a in 0:50, b in 51:100
    pl_pdf = plot([x -> pdf(Uniform(a,b), x)], a, b)
    pl_cdf = plot([x -> cdf(Uniform(a,b), x)], a, b)
    vstack(pl_pdf, pl_cdf)
end   
```

### supported on semi-infinite intervals, usually [0,∞)

#### Chi-squared distribution

The chi-squared distribution, which is the sum of the squares of n independent Gaussian random variables.

$$ f(x; k) = \frac{x^{k/2 - 1} e^{-x/2}}{2^{k/2} \Gamma(k/2)}, \quad x > 0.$$

```julia
chisq_values = rand(Chisq(2),5)
```

```julia
@manipulate for v in 1:20
    pl_pdf = plot([x -> pdf(Chisq(v), x)], 0, 10)
    pl_cdf = plot([x -> cdf(Chisq(v), x)], 0, 10)
    vstack(pl_pdf, pl_cdf)
end   
```

#### Exponential distribution

The exponential distribution, which describes the time between consecutive rare random events in a process with no memory

$$f(x; \theta) = \frac{1}{\theta} e^{-\frac{x}{\theta}}, \quad x > 0$$

```julia
exponential_values = rand(Exponential(2),5)
```

```julia
@manipulate for θ in .1:.1:2
    pl_pdf = plot([x -> pdf(Exponential(θ), x)], 0, 10)
    pl_cdf = plot([x -> cdf(Exponential(θ), x)], 0, 10)
    vstack(pl_pdf, pl_cdf)
end   
```

#### F-distribution

The F-distribution, which is the distribution of the ratio of two (normalized) chi-squared-distributed random variables, used in the analysis of variance. 

$$ f(x; \nu_1, \nu_2) = \frac{1}{x B(\nu_1/2, \nu_2/2)}
\sqrt{\frac{(\nu_1 x)^{\nu_1} \cdot \nu_2^{\nu_2}}{(\nu_1 x + \nu_2)^{\nu_1 + \nu_2}}}, \quad x>0$$

```julia
fdist_values = rand(FDist(2, 1),5)
```

```julia
@manipulate for ν1 in 1:10, ν2 in 1:10
    pl_pdf = plot([x -> pdf(FDist(ν1, ν2), x)], 0, 20)
    pl_cdf = plot([x -> cdf(FDist(ν1, ν2), x)], 0, 20)
    vstack(pl_pdf, pl_cdf)
end   
```

#### Normal distribution

The normal distribution, also called the Gaussian or the bell curve. It is ubiquitous in nature and statistics due to the central limit theorem: every variable that can be modelled as a sum of many small independent, identically distributed variables with finite mean and variance is approximately normal.

$$ f(x; \mu, \sigma) = \frac{1}{\sqrt{2 \pi \sigma^2}}
\exp \left( - \frac{(x - \mu)^2}{2 \sigma^2} \right)$$

```julia
normal_values = rand(Normal(2, 1),5)
```

```julia
@manipulate for μ in -5:10, σ in 1:5
    pl_pdf = plot([x -> pdf(Normal(μ,σ), x)], μ-10, μ+10)
    pl_cdf = plot([x -> cdf(Normal(μ,σ), x)], μ-10, μ+10)
    vstack(pl_pdf, pl_cdf)
end   
```

#### Student's t-distribution

Student's t-distribution, useful for estimating unknown means of Gaussian populations.

$$f(x; d) = \frac{1}{\sqrt{d} B(1/2, d/2)}
\left( 1 + \frac{x^2}{d} \right)^{-\frac{d + 1}{2}}$$

```julia
tdist_values = rand(TDist(10),5)
```

```julia
@manipulate for d in 1:50
    pl_pdf = plot([x -> pdf(TDist(d), x)], -10, 10)
    pl_cdf = plot([x -> cdf(TDist(d), x)], -10, 10)
    vstack(pl_pdf, pl_cdf)
end   
```
