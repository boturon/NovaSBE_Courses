```julia
using Gadfly; using LinearAlgebra;
```

```julia
k_ss(α, A, s, n, δ, γ) = ((s*A)/(n+δ+γ))^(1/(1-α))
y_ss(α, A, s, n, δ, γ) = A*k_ss(α, A, s, n, δ, γ)^α;
```

<br>

```julia
function ex1c()
    t_crisis = 40
    t = 200
    α = 0.5; A = 2; s = 0.1 ; n = 0.03; δ = 0.07; γ = 0;
    values = zeros(t,3); values[1,:] = [k_ss(α, A, s, n, δ, γ) y_ss(α, A, s, n, δ, γ) (1-s)*y_ss(α, A, s, n, δ, γ)]
    #values = zeros(t,3); values[1,:] = [(6/2)^(1/0.5) 6 (1-s)*6];
    for i in 2:t
        if i > t_crisis
            s = 0.3
            kt = (s*values[i-1,2]+(1-δ)*values[i-1,1])/(1+n+(1+n)*γ)
            values[i,:] = [kt A*kt^α (1-s)*A*kt^α]
        else
            kt = (s*values[i-1,2]+(1-δ)*values[i-1,1])/(1+n+(1+n)*γ)
            values[i,:] = [kt A*kt^α (1-s)*A*kt^α]
        end
    end
    p = hstack(
            plot(
                layer(x=1:t, y=values[:,1], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[t_crisis t_crisis], y=[0 values[t_crisis,1]], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Capital per capita, k")), 
            plot(
                layer(x=1:t, y=values[:,2], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[t_crisis t_crisis], y=[0 values[t_crisis,2]], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Output per capita, y")), 
            plot(
                layer(x=1:t, y=values[:,3], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[t_crisis t_crisis], y=[0 values[t_crisis-1,3]], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Consumption per capita, c")))
    draw(SVG("solow/1c.svg", 21cm, 7cm), p)
    return draw(SVG(21cm, 7cm), p)
end;
```

```julia
function ex3b()
    t_crisis = 40
    t = 300
    α = 0.5; A = 2; s = 0.25; n = 0.02; δ = 0.03; γ = 0;
    #values = zeros(t,3); values[1,:] = [k_ss(α, A, s, n, δ, γ) y_ss(α, A, s, n, δ, γ) (1-s)*y_ss(α, A, s, n, δ, γ)]
    values = zeros(t,3); values[1,:] = [(6/2)^(1/0.5) 6 (1-s)*6];
    for i in 2:t
        #if i == t_crisis
        #    kt = values[i-1,1]/2
        #    values[i,:] = [kt A*kt^α (1-s)*A*kt^α]
        #else
            kt = (s*values[i-1,2]+(1-δ)*values[i-1,1])/(1+n+(1+n)*γ)
            values[i,:] = [kt A*kt^α (1-s)*A*kt^α]
        #end
    end
    p1 = hstack(
            plot(
                layer(x=1:t, y=values[:,1], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[1 t], y=[100 100], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Capital per capita, k")), 
            plot(
                layer(x=1:t, y=values[:,2], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[1 t], y=[20 20], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Output per capita, y")), 
            plot(
                layer(x=1:t, y=values[:,3], Geom.line, Theme(default_color=colorant"black")),
                layer(x=[1 t], y=[15 15], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Consumption per capita, c")))
    
    p2 = hstack(
            plot(
                layer(x=2:t, y=(log.(values[2:end,1]).-log.(values[1:end-1,1])), Geom.line, Theme(default_color=colorant"black")),
                layer(x=[2 t], y=[0 0], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Growth of k")), 
            plot(
            layer(x=2:t, y=(log.(values[2:end,2]).-log.(values[1:end-1,2])), Geom.line, Theme(default_color=colorant"black")),
                layer(x=[2 t], y=[0 0], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Growth of y")), 
            plot(
            layer(x=2:t, y=(log.(values[2:end,3]).-log.(values[1:end-1,3])), Geom.line, Theme(default_color=colorant"black")),
                layer(x=[2 t], y=[0 0], Geom.line,  style(line_style=[:dot], default_color=colorant"black")),
                Guide.xlabel("Time"), Guide.ylabel("Growth of c")))
    
    o = vstack(p1,p2)
    
    draw(SVG("solow/3b.svg", 21cm, 14cm), o)
    return draw(SVG(21cm, 14cm), o)
end;
```
