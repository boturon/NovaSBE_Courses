```python
#data = pd.read_csv("/home/boturon/Desktop/Stocks.csv", sep=";", index_col=0)
#data = pd.read_excel("./Thesis_AA (1).xlsx", sheet_name="offline")
data = pd.read_csv("~/Desktop/AA/Thesis_AA.csv", index_col=0)
data.index = pd.to_datetime(data.index)
data[data==" "] = np.nan
data = data.dropna(axis=1, how='all')
data = data.astype(np.float)
```

<br>

```python
returns = np.log(data).diff().dropna(axis=0)
```

```python
returns.head(10)
```

```python
returns.to_csv("~/Desktop/AA/data_returns.csv", index=False)
```

<br>

```python
betas = pd.DataFrame(columns=["E1", "E2", "beta", "pvalue", "intercept", "rvalue", "stderr"])
betas["E1"], betas["E2"] = zip(*itertools.combinations(returns.columns, 2))
for index, row in betas.iterrows():
    row["beta"], row["intercept"], row["rvalue"], row["pvalue"], row["stderr"] = stats.linregress(returns[row["E2"]],returns[row["E1"]])
```

```python
betas[betas.pvalue<0.05].head(20)
```

```python
betas[betas.pvalue<0.05].to_csv("./data_betas.csv", index=False)
```
