```python
# Solution
# Define a function named "factorial". The function must calculate the factorial of a variable "n".
def factorial(n):
    result = 1
    for val in range(n):
        result*=val+1
    return result

# Create a new variable "sparrow_count" containing the number of characters present in the string "Jack Sparrow"
sparrow_count = len('Jack Sparrow')

# Demonstrate the functioning of the function by running factorial(sparrow_count)
factorial(sparrow_count)
```

### 2.13. Create a function called createGroups that receives a dataset and the n_elements for group formation.
Rules to create groups:
1. Use the df_pokemon dataset
2. the number of elements in each group is set by n_elements (x valores)
3. if the number of students is not enough to create a final group, the remaining students must be put in a random selected group (x valores)



```python
# Solution 1
def createGroups(dataset, elements):    
    
    df = df_pokemon.sort_values(by='Total', ascending=False)
    
    n_groups = int(len(dataset)/elements)
    group = n_groups
    counter = 0
    #Create groups
    for i in df.index:
        if(counter == elements):
            group = group - 1
            counter = 0
        df.at[i, 'group'] = group
        counter = counter + 1

    #replace elements without group
    prev_group = 0
    for i in df.index:
        if (df.at[i, 'group'] == 0):
            replace_group = randint(1, n_groups)
            while (replace_group == prev_group):
                replace_group = randint(1, n_groups)
            prev_group = replace_group
            df.at[i, 'group'] = replace_group
    return df

createGroups(df_pokemon, 3)


# solution 2
def mkgroups(df, n_elements):
    df = df.sort_values(by='Total')
    n_groups = int(len(df)/n_elements)
    # make groups dict
    groups = {}
    for group in range(n_groups):
        groups[str(group)] = []
    # function to assign groups
    def assign_group(row):
        for group in range(n_groups): 
            if len(groups[str(group)]) < 3:
                groups[str(group)].append(row['nbr'])
                return group
        return randint(1, n_groups) # return breaks the function, it only reaches this point if all groups are filled
        
    df['group'] = df.apply(assign_group, axis=1)
    return df, groups # also returns the groups in a dictionary

print(mkgroups(df_pokemon, 3)[1])
mkgroups(df_pokemon, 3)[0]
```
