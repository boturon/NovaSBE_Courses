```python
import pandas as pd
import matplotlib
import pandas_profiling
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

%matplotlib inline
set_matplotlib_formats('retina')
```

```python
df = pd.read_csv('Crime1.csv')
df.head()
```

```python
df.describe()
df.info()
pandas_profiling.ProfileReport(df)
```

```python
df.groupby('Category').size().to_frame().rename(columns={0:'count'}).sort_values('count', ascending=False).plot.bar()
plt.show()
```

```python
crimes = ['LARCENY/THEFT','VEHICLE THEFT','ASSAULT','ROBBERY','KIDNAPPING']

# filter dataframe to only show the crimes we want to analyse
df = df[df.Category.isin(crimes)]
```

```python
# bar chart with count of crimes per district
df.groupby('PdDistrict')\
    .size()\
    .to_frame()\
    .rename(columns={0:'count'})\
    .sort_values('count', ascending=False)\
    .plot.bar()

plt.show()
```

```python
df.Dates = pd.to_datetime(df.Dates)
df['Hour'] = df.Dates.dt.hour

df2 = df.groupby('Hour').size().to_frame().rename(columns={0:'crime_count'})#.reset_index()

df2.plot.bar()
plt.show()
```

```python
df3 = df[df.Category.isin(['LARCENY/THEFT','VEHICLE THEFT','ASSAULT','ROBBERY','KIDNAPPING'])]\
    .groupby(['Hour', 'Category']).size().to_frame()\
    .rename(columns={0:'crime_count'})#.reset_index()

df3.reset_index().pivot(index='Hour', columns='Category', values='crime_count').plot.bar(stacked=True)
```

```python
df = df[df['PdDistrict'].isin(['NORTHERN', 'SOUTHERN'])]
```

```python
def resolution_edit(val):
    if val=='NONE':
        return 'Never found suspect'
    else:
        return 'Action was taken'

df['PdSuccess'] = df.Resolution.apply(resolution_edit)
```

```python
v = df.groupby(['PdSuccess', 'PdDistrict']).size().to_frame().reset_index().rename(columns={0:'count'})#.plot.bar()

v.pivot(index='PdSuccess', columns='PdDistrict', values='count').plot.bar()
plt.show()
```

```python
v = df.groupby(['PdSuccess', 'Category']).size().to_frame().reset_index().rename(columns={0:'count'})#.plot.bar()
v[v.Category.isin(['ROBBERY','ASSAULT','KIDNAPPING','VEHICLE THEFT'])].pivot(index='PdSuccess', columns='Category', values='count').plot.bar()
plt.show()
```

```python
v = df[df.Category.isin(['ROBBERY','ASSAULT','KIDNAPPING','VEHICLE THEFT'])]\
    .groupby(['PdSuccess', 'Category', 'PdDistrict'])\
    .size()\
    .to_frame()\
    .reset_index()\
    .rename(columns={0:'count'})\
    .append({'PdSuccess':'Action was taken', 'Category':'KIDNAPPING', 'PdDistrict':'SOUTHERN', 'count':0}, ignore_index=True)


fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
a=0
for i in list(v.PdDistrict.unique()):
    b = v[v.PdDistrict==i].pivot(index='PdSuccess', 
                                 columns='Category', 
                                 values='count').plot.bar(ax=axes[a])
    b.set_title(i)
    a+=1










```
