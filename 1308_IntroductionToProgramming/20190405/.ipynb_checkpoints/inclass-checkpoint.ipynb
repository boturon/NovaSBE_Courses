### Part 1: setting up functions and utilities

```python

def reader(directory):
    """
    Reads a dataframe
    """
    df = pd.read_csv(directory)
    return df
    
df = reader('Crime1.csv')

df.head()
```

```python

def preprocess(df, show_report=True):
    """
    Edits the dataframe according to the analyses wanted.
    
    Params:
        - df: dataframe to be preprocessed
        - show_report: Whether to show basic info on Jupyter Notebook or not. Defaults to True
        
    Returns a preprocessed dataframe
    """
    # drop duplicate rows
    df2 = df.drop_duplicates()

    # convert Dates to datetime format
    df2.Dates = pd.to_datetime(df2.Dates)
    df2['Hour'] = df2.Dates.dt.hour
    
    if show_report:
        return df2, pandas_profiling.ProfileReport(df2)
    else:
        return df2

new_df, report = preprocess(df)


#new_df
```

```python
def mk_bar_chart(df, group_cols, p_index=None, p_columns=None, p_values=None, sort_vals_col='count', method='groupby', **kwargs):
    """
    method:
     - groupby
     - groupby_sorted
     - pivot
    """
    set_matplotlib_formats('retina')
    
    to_plot_df = df.groupby(group_cols)\
                    .size()\
                    .to_frame()\
                    .rename(columns={0:'count'})
    
    if method == 'groupby_sorted':
        to_plot_df.sort_values(sort_vals_col, ascending=False)\
            .plot.bar(**kwargs)
        
    elif method == 'groupby':
        to_plot_df.plot.bar(**kwargs)
    
    elif method == 'pivot':
        to_plot_df.reset_index()\
            .pivot(index=p_index, columns=p_columns, values=p_values)\
            .plot.bar(**kwargs)
    else:
        raise TypeError('%s is not a valid method. Valid methods are [groupby, groupby_sorted, pivot].' % method)
        
    plt.show()
```

```python
mk_bar_chart(new_df, 'Category', method='groupby_sdasrted', title='this is a nice title')
```
