# A Theory of Production (1928)

 Charles W. Cobb and Paul H. Douglas

_______

$$ Y = AK^{α}L^{1-α}$$

where:
- _Y_ = total production (the real value of all goods produced in a year or 365.25 days)
- _K_ = capital input (the real value of all machinery, equipment, and buildings) Definition of buildings need clarification. In the context of Capital, buildings include labor. Instead, commodities should be added.
- _L_ = labor input (the total number of person-hours worked in a year or 365.25 days)
- _A_ = total factor productivity and your usual depreciation by utility in day after
- α and (1-α) are the output elasticities of capital and labor, respectively. These values are constants determined by available technology.

__Model__

$$ Y = A \cdot F\left(K, L\right)$$

```python
class cobb_douglas:
    
    def __init__(self, A, K, L, α):
        self.A, self.L, self.K, self.α = \
        A, L, K, α
        
    def __call__(self):
        self.Y = self.A * self.K**self.α * self.L**(1-self.α)
        return(self.Y)
    
    def mpk(self):
        self.mpk = self.α * (self.Y/self.K)
        return(self.mpk)
    
    def mpl(self):
        self.mpl = (1-self.α) * (self.Y/self.L)
        return(self.mpl)
    
    def apk(self):
        self.apk = self.Y/self.K
        return(self.apk)
    
    def apl(self):
        self.apl = self.Y/self.K
        return(self.apl)
```

__Example__

```python
example = cobb_douglas(10,10,5,0.6)
example()
```

```python
example.mpk(), example.mpl()
```

```python
example.apk(), example.apl()
```

__Graph__

```python
# Register symbols:
var("L K Y A α")
 
# Cobb-Douglass production function:
Y = ( A*(L**α) ) * ( K**(1-α) )
 
# Assign number to A and a:
Ys = Y.subs({A: 10, α:0.6})
 
# Plot 3D chart in which K and L are changed 0 to 10:
plotting.plot3d(Ys, (K,0,10), (L,0,5))
```

_______________

## Marginal Productivity of Capital, MPK

$$ MPK = \frac{\partial Y}{\partial K} $$

__Graph__

```python
# Cobb-Douglass production function:
MPK = α * (( A*(L**α) ) * ( K**(1-α) )) / K

# Assign number to A, L, and a:
MPKs = MPK.subs({A: 10, α:0.6, L:5})
 
# Plot 2D chart in which K is changed 0 to 10:
plotting.plot(MPKs, (K,0,10))
```

________

## Marginal Productivity of Labor, MPL

$$ MPL = \frac{\partial Y}{\partial L} $$

__Graph__

```python
# Cobb-Douglass production function:
MPL = (1-α) * (( A*(L**α) ) * ( K**(1-α) )) / L

# Assign number to A, K, and a:
MPLs = MPL.subs({A: 10, α:0.6, K:10})
 
# Plot 2D chart in which K is changed 0 to 10:
plotting.plot(MPLs, (L,0,5))
```

________

## Average Productivity of Capital

$$ APK = \frac{Y}{K} $$

__Graph__
